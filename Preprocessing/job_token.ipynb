```python
from nltk.tokenize import word_tokenize, sent_tokenize

def safe_word_tokenize(text):
    try:
        return word_tokenize(text)
    except LookupError:
        # Fallback: simple whitespace split if tokenizer resource missing
        return text.split()

df['tokenised'] = df['text_cleaned'].apply(lambda x: safe_word_tokenize(x))
```
